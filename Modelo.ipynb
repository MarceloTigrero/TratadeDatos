```python
Folderbydata = 'data/S1'  
files = os.listdir( Folderbydata )
#files

# me quedo con los archivos que son .csv
files = [file for file in files if file.endswith(".csv")]
files
```

```python
# creo un diccionario para guardar los datos de los archivos
Variables = ['Ax','Ay','Az','Gx','Gy','Gz']
S1 = {}
for file in files:
    name = file.split('.')[0]
    tarea = int(name.split('S')[0][-1])
    rep = int(name.split('R')[1])
    #print(tarea, rep)
    if tarea not in S1:
        S1[tarea] = {}
        S1[tarea][rep] = {}
        for var in Variables:
            S1[tarea][rep][var] = []
        for row in open (Folderbydata + '/' + file):
            row = row.split(',')
            for i in range(len(row)):
                try:
                    S1[tarea][rep][Variables[i]].append(float(row[i]))
                except:
                    print(row[i])
                    pass
    else:
        if rep not in S1[tarea]:
            S1[tarea][rep] = {}
            for var in Variables:
                S1[tarea][rep][var] = []
            for row in open (Folderbydata + '/' + file):
                row = row.split(',')
                for i in range(len(row)):
                    try:
                        S1[tarea][rep][Variables[i]].append(float(row[i]))
                    except:
                        print(row[i])
                        pass
        else:
            for row in open (Folderbydata + '/' + file):
                row = row.split(',')
                for i in range(len(row)):
                    try:
                        S1[tarea][rep][Variables[i]].append(float(row[i]))
                    except:
                        print(row[i])
                        pass

print(S1)
```

```python
# datos de entrada & salida
datosinput = []
datosoutput = []
for tarea in S1:
    for rep in S1[tarea]:
        lista = []
        for var in Variables:
            lista.append(np.array(S1[tarea][rep][var]))
        # valido si hay datos  
        for i in range(len(lista)):
            if len(lista[i]) > 0:
                datosinput.append(np.array(lista))
                datosoutput.append(tarea)
datosinput = np.array(datosinput)
datosoutput = np.array(datosoutput)
```

```python
inputlayer = tf.keras.layers.Dense(units=6, input_shape=(x_train.shape[1],), activation='relu') 
#oculta2 = tf.keras.layers.Dense(units=500, activation='relu')#datosinput.shape[0]
oculta = tf.keras.layers.Dense(units=7, activation='softmax') 
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([inputlayer, oculta, salida])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.01),
    #loss='mean_squared_error',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(x_train, y_train, epochs=1000,verbose=1 )#
print("Modelo entrenado!")
```
